```python
from scipy.io import loadmat
x = loadmat('teapot.mat')
import numpy as np
```

```python
a = []
b = []
c = []
for points in x['verts']:
    a.append(points[0])
    b.append(points[1])
    c.append(points[2])
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(a,b,c, c='b', marker='*')
```

```python
### Your implementation comes here
homo = np.ones((len(a)))
points = np.array([a,b,c,homo])
degIncrement= 10
cameraMat = np.array([[1,0,0],[0,0,1]])
plots = []
# print(points.shape)
W = np.zeros((72,306))

deg = 0
counter = 0
for i in range(36):
    R = np.array([[np.cos(np.radians(deg)), -(np.sin(np.radians(deg))),
                   0, 0],[np.sin(np.radians(deg)), np.cos(np.radians(deg)), 0, 0],
                  [0,      0,    1, 0]])
    
    img = np.matmul(cameraMat,np.matmul(R,points))

    plots.append(img)                             
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.scatter(img[0],img[1], c='g', marker='*')
    ax.set_title(str(deg) + " Degrees")
    deg +=10      
    W[counter] = img[0]
    W[counter+1] = img[1]
    counter+=2
    

print (len(plots))
```

### Task 2
#### Use the data generated in previous task to generate a W matrix, and use that as input in Tomasi Kanade Factorization method to compute the 3D structure

```python
### Your implementation comes here
U, S, Vt = np.linalg.svd(W);
# print(U)
# print(U.shape,S.shape,Vt.shape)

R_hat = U[:,0:3]
# print(R_hat.shape)

V_hat = Vt[0:3,:]
# print(V_hat.shape)

S_hat = np.array([[S[0], S[1], S[2]],
                 [S[3], S[4], S[5]],
                 [S[6], S[7], S[8]]])
# S_hat = S[0:3,0:3]
# print(S_hat.shape)

S_hat = np.matmul(S_hat, V_hat)
print(S_hat.shape)

print(R_hat.shape,S_hat.shape,V_hat.shape)

c = 0
c2 = 0

A= np.zeros((36*3, 6))
B = np.zeros((3*36,1))

for i in range(36):
    j=c;
    j2=c+1;
    r1 = R_hat[j][0]
    r2 = R_hat[j][1]
    r3 = R_hat[j][2]
    
    r4 = R_hat[j2][0]
    r5 = R_hat[j2][1]
    r6 = R_hat[j2][2]
    c+=2
    
    A[c2]   = [(r1*r1), (2*r1*r2), (2*r1*r3), (r1*r2), (2*r2*r3), (r3*r3)]
    A[c2+1] = [(r4*r4), (2*r4*r5), (2*r4*r6), (r5*r5), (2*r5*r6), (r6*r6)]
    A[c2+2] = [(r1*r4), (2*r1*r5), (2*r1*r6), (r2*r5), (2*r2*r6), (r6*r6)]
    B[c2]   = [1]
    B[c2+1] = [1]
    B[c2+2] = [0]
    
    c2+=3
    

#     r4 = row1(2,1); r5 = row1(2,2); r6 = row1(2,3);
#     A = [A; r1*r1, 2*r1*r2, 2*r1*r3, r2*r2, 2*r2*r3, r3*r3];
#     A = [A; r4*r4, 2*r4*r5, 2*r4*r6, r5*r5, 2*r5*r6, r6*r6];
#     A = [A; r1*r4, 2*r1*r5, 2*r1*r6, r2*r5, 2*r2*r6, r6*r6];
#     B = [B, 1, 1, 0];
# end

X = np.matmul(np.linalg.inv(np.matmul(A.T,A)),(np.matmul(A.T,B)))
# print(X)

Q= np.zeros((3,3))
Q[0] = [X[0], X[1], X[2]]
Q[1] = [X[2], X[3], X[4]]
Q[2] = [X[2], X[4], X[5]]

# print(Q)
Q_hat = np.linalg.cholesky(Q)
# print(Q_hat)
S = np.matmul(np.linalg.inv(Q_hat),S_hat)
# print(S.shape)

R = np.matmul(R_hat, Q_hat)
# print(R.shape)
W_new = np.matmul(R,S)
print(W_new)
print(W_new.shape)
```
