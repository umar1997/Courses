```python
# Group By
df = pd.read_csv('./CSV Files/census.csv')
df = df[df['SUMLEV']==50]
df.head()
######################################################
for state in df['STNAME'].unique():
    avg = np.average(df[df['STNAME']==state].dropna()['CENSUS2010POP'])
    print("Countries in state"+state+"has an avg population of"+str(avg))
######################################################
# Or can be done by the Group By method which comines all the same entries together
# Group takes all the unique ids or whatever group you have chosen
# Frame is the data frame of that group
for group, frame in df.groupby('STNAME'):
#     print(frame[['REGION','COUNTY']])
    avg= np.average(frame['CENSUS2010POP'])
    print("Countries in state "+ group +"has an avg population of"+str(avg))
######################################################  
# group by a function by setting the column you need as index
df = df.set_index('STNAME')
def func(item):
    if(item[0] < 'M'):
        return 0
    if(item[0] < 'V'):
        return 1
    return 2
for group, frame in df.groupby(func):
    print("There are" + str(len(frame)) + " records in group " str(group))
######################################################
# In agg you pass in a dictionary name the columns 
# you want and after the : the function you want to apply
df.groupby('STNAME').agg({'CENSUS2010POP':np.average})

# NOTE: V good examples of group by in customers.ipynb with explanation
```

```python
# Scales
# 1) Ratio Scale +-/* 
# 2) Interval Scale (O means something e.g in temp or an a compass)
# 3) Ordinal Scale (Grades)
# 4) Nominal Scale(Categorical Data) (Colour)

sdf = pd.DataFrame(['A+','A','A-','B+','B','B-','C+','C','C-','D+','D'],
                  index = ['excellent','excellent','excellent','good','good','good','ok','ok','ok','poor','poor'])
sdf.rename(columns = {0:'Grades'}, inplace= True)
grades = sdf['Grades'].astype('category',
                     categories = ['D','D+','C-','C','C+','B-','B','B+','A-','A','A+'],
                     ordered = True)
print(grades)
#######################################
print(grades >'C')
#######################################
# The astype function changes nominal to ordinal data
s = pd.Series(['Low', 'Low', 'High', 'Medium', 'Low', 'High', 'Low'])
s.astype('category', categories=['Low', 'Medium', 'High'], ordered=True)
#######################################
# The CUT is used for changing interval or ratio data into categorial data 
size  = pd.Series([168, 180, 174, 190, 170, 185, 179, 181, 175, 169, 182, 177, 180, 171])
sizeCategory = pd.cut(size, 3, labels=['Small', 'Medium', 'Large'])
sizeCategory.head()
```

```python
# DATE FUNCTIONALITY
import pandas as pd
import numpy as np

# TimeStamp
# the index of timestaamp is datetime index
print(pd.Timestamp('9/1/2016 10:05AM'))
print("***********A************")
# Period
print(pd.Period('1/2016')) # MM/YYYY
print(pd.Period('3/5/2016')) # MM/DD/YYYY
print("***********B*************")
# DateTime Index
t1 = pd.Series(list('abc'),[pd.Timestamp('2016-09-01'),pd.Timestamp('2016-09-02'),pd.Timestamp('2016-09-03')])
print(t1) # YYYY-MM-DD
print(type(t1.index))
print("************C************")
# PeriodIndex
t2 = pd.Series(list('def'),[pd.Period('2016-09'),pd.Period('2016-10'),pd.Period('2016-11')])
print(t2)
print(t2.index)
print("************D************")
# Converting to Date Time
d1 = ['2nd June 2013', 'Aug 29 2014', '2015-06-26','7/12/16']
x = np.random.randint(10,100,(4,2))
print(x)
t3 = pd.DataFrame(x, index = d1, columns = ['First', 'Second'])
print(t3)
t3.index = pd.to_datetime(t3.index)
print(t3)
print("*************E***********")
# Time Deltas (The differences within time)
print(pd.Timestamp('9/2/2016 8:10AM') + pd.Timedelta('12D 3H'))
# What if we want 9 dates for each sunday starting in Oct 2016
dates = pd.date_range('10-01-2016', periods=9, freq = '2W-SUN')
print(dates)
print("************F************")
df = pd.DataFrame({'Count 1': np.random.randint(10,100,9),
                  'Count 2': np.random.randint(10,100,9)},
                  index = dates)
print(df)
print(df.index.weekday_name)
print("************G************")
print(df.diff()) # This gives the difference of values from the value above it
print("************H************")
print(df.resample('M').mean()) # Find each month ka mean
print(df['2017']) # Find those values which are in 2017
print("************I************")
# Plotting the Time Series
import matplotlib.pyplot as plt; plt.rcdefaults()
df.plot()
```
